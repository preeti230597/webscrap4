# 1.. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
#importing all libraries
```

```python
driver=webdriver.Chrome("C:\\Users\\Preeti kumari\\chromedriver.exe")
#opening the driver
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
#sending the link
```

```python
title=driver.find_elements_by_tag_name("tbody")
p=title[0]
```

```python
rank=[]
r=p.find_elements_by_xpath("tr/td[1]")
for i in r:
    if i.text is None :
        rank.append("--") 
    else:
        rank.append(i.text.replace('\n'," "))   
#scrapping the rank
```

```python
name=[]
n=p.find_elements_by_xpath("tr/td[2]/a")
for i in n:
    if i.text is None :
        name.append("--") 
    else:
        name.append(i.text.replace('\n'," "))   
#scrapping video name
```

```python
uploader=[]
u=p.find_elements_by_xpath("tr/td[3]")
for i in u:
    if i.text is None :
        uploader.append("--") 
    else:
        uploader.append(i.text.replace('\n'," "))   
#scrapping youtuber name
```

```python
views=[]
v=p.find_elements_by_xpath("tr/td[4]")
for i in v:
    if i.text is None :
        views.append("--") 
    else:
        views.append(i.text.replace('\n'," "))   
len(views)
#scrapping views on yt
```

```python
upload_date=[]
ud=p.find_elements_by_xpath("tr/td[5]")
for i in ud:
    if i.text is None :
        upload_date.append("--") 
    else:
        upload_date.append(i.text.replace('\n'," "))   
len(upload_date)
#scrapping the upload date
```

```python
data=list(zip(rank,name,uploader,upload_date,views))
#saving in list data
```

```python
df=pd.DataFrame(data,columns=["Rank","Name","Uploader","Upload_date","Views(in Billion)"])
#creating a df
```

```python
df.set_index(['Rank'], inplace = True) 
#setting the index as rank
```

```python
name=p.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
len(name)
#finding the total number of trending repository 
```

```python
rep_name=[]
for i in name:
    if i.text is None:
        rep_name.append("-")
    else:
        rep_name.append(i.text)
#Scrapping repository name
```

```python
desc=[]
d=driver.find_elements_by_xpath("//article[@class='Box-row']")
for i in range(0,25):
    dd=d[i]
    k=dd.find_elements_by_tag_name('p')
    if len(k)==0:
        desc.append('-')
    else:
        m=k[0]
        desc.append(m.text)
#scrapping description of the repository
```

```python
contri_count=[]
for i in range(0,25):
    kk=driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']")
    k=kk[i].find_elements_by_tag_name('a')
    if len(k)==0:
        contri_count.append('-')
    else:
        p=k[1]
        contri_count.append(p.text)
#scrapping the number of contributors
```

```python
#creating a df
df_git=pd.DataFrame({"Repository Name":rep_name,"Language":lang,"Description":desc,"No of Contributors":contri_count})
```
